# 4.9 Running, Restarting, and Interrupting Jupyter Kernels

This section demonstrates how to control a Jupyter kernel intentionally. 
We will:
1. Run cells and observe state.
2. Restart the kernel to clear variables.
3. Interrupt a deliberately long-running cell.
4. Understand the difference between restarting and interrupting.
</VSCode.Cell>
<VSCode.Cell language="markdown">
## 1. Running Cells and Understanding Execution Order
Let's assign a variable and print it. This stores it in the kernel's memory.
</VSCode.Cell>
<VSCode.Cell language="python">
# Run this cell first
my_var = "Hello, Kernel!"
print(my_var)
</VSCode.Cell>
<VSCode.Cell language="python">
# This cell remembers 'my_var' because it was stored in the active kernel
print(f"The kernel still remembers: {my_var}")
</VSCode.Cell>
<VSCode.Cell language="markdown">
## 2. Restarting Kernel
If you click **Restart Kernel**, the memory is wiped out. 
Try clicking restart from the Jupyter menu/bar, and then try running the cell above *without* running the first one. You'll get a `NameError` because `my_var` no longer exists.
</VSCode.Cell>
<VSCode.Cell language="markdown">
## 3. Interrupting Execution
Sometimes a cell gets stuck in an infinite loop or takes too long. 
Run the cell below, then click **Interrupt Kernel** (the stop button ⏹️) before it finishes.
</VSCode.Cell>
<VSCode.Cell language="python">
import time

print("Starting a 20-second sleep... Click Interrupt (⏹️) quickly!")
for i in range(20):
    time.sleep(1)
    print(f"Slept {i+1} seconds...")
print("Finished!")
</VSCode.Cell>
<VSCode.Cell language="markdown">
## 4. Recognizing When to Restart vs. Interrupt

- **Interrupt:** Use when a cell is taking too long to run (e.g., infinite loop, big data download). It stops the current execution but **keeps** all your existing variables in memory.
- **Restart:** Use when your notebook is behaving weirdly, variables are corrupted, or you want to start fresh to make sure your code runs from top to bottom perfectly. It **destroys** all temporary memory/variables.

```python
## Data Analysis

The dataset contains student feedback across multiple areas such as teaching, course content, examination, lab work, library facilities, and extracurricular activities.

Each category is rated between -1 (negative), 0 (neutral), and 1 (positive).
```

```python
import pandas as pd

df = pd.read_excel("finalDataset0.2.xlsx")  # or CSV if you prefer
df.head()
```

```python
df.columns = [
    'teaching_rating', 'teaching_comment',
    'course_rating', 'course_comment',
    'exam_rating', 'exam_comment',
    'lab_rating', 'lab_comment',
    'library_rating', 'library_comment',
    'extra_rating', 'extra_comment'
]

df.head()
```

```python
df.info()
```

```python
def normalize(x):
    return (x + 1) / 2

rating_cols = [
    'teaching_rating','course_rating','exam_rating',
    'lab_rating','library_rating','extra_rating'
]

for col in rating_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df[col].fillna(0)
    df[col] = df[col].apply(normalize)
```

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

comment_cols = [
    'teaching_comment','course_comment','exam_comment',
    'lab_comment','library_comment','extra_comment'
]

for col in comment_cols:
    df[col] = df[col].apply(clean_text)
```

```python
from textblob import TextBlob

for col in comment_cols:
    df[col + '_sentiment'] = df[col].apply(
        lambda x: TextBlob(x).sentiment.polarity
    )
```

```python
df['teaching_score'] = df['teaching_rating']*0.6 + df['teaching_comment_sentiment']*0.4
df['course_score'] = df['course_rating']*0.6 + df['course_comment_sentiment']*0.4
df['exam_score'] = df['exam_rating']*0.6 + df['exam_comment_sentiment']*0.4
df['lab_score'] = df['lab_rating']*0.6 + df['lab_comment_sentiment']*0.4
df['library_score'] = df['library_rating']*0.6 + df['library_comment_sentiment']*0.4
df['extra_score'] = df['extra_rating']*0.6 + df['extra_comment_sentiment']*0.4
```

```python
df['overall_score'] = df[
    ['teaching_score','course_score','exam_score',
     'lab_score','library_score','extra_score']
].mean(axis=1)

df['overall_score'].mean()
```

```python
kpi = df[
    ['teaching_score','course_score','exam_score',
     'lab_score','library_score','extra_score']
].mean().sort_values()

kpi
```

```python
import matplotlib.pyplot as plt

kpi.plot(kind='bar')
plt.title("Student Satisfaction by Category")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.show()
```
