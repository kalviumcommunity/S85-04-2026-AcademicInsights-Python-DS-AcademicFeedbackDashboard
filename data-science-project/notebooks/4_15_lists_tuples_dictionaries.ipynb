# 4.15 — Working with Python Lists, Tuples, and Dictionaries

This notebook covers Python collection fundamentals: `list`, `tuple`, and `dict`.

## 1) Working with Python Lists (ordered, mutable)

```python
# Create a list
courses = ['Python', 'SQL', 'Excel', 'Power BI']
print('Original list:', courses)

# Access by index
print('First item:', courses[0])
print('Last item:', courses[-1])

# Modify an element
courses[2] = 'Advanced Excel'
print('After modification:', courses)

# Add and remove elements
courses.append('Statistics')
removed = courses.pop(1)
print('After append and pop:', courses)
print('Removed item:', removed)
```

```python
# Iterate over list items
print('Iterating over list:')
for i, item in enumerate(courses, start=1):
    print(f'{i}. {item}')
```

## 2) Working with Python Tuples (ordered, immutable)

```python
# Create a tuple
student_profile = ('Pranav', 21, 'Data Science')
print('Tuple:', student_profile)

# Access by index
print('Name:', student_profile[0])
print('Track:', student_profile[2])
```

```python
# Demonstrate immutability safely
try:
    student_profile[1] = 22
except TypeError as e:
    print('Tuple immutability check:', e)
```

## 3) Working with Python Dictionaries (key-value, mutable)

```python
# Create a dictionary
student = {
    'name': 'Pranav',
    'age': 21,
    'course': 'Data Science',
    'active': True
}
print('Original dictionary:', student)

# Access by key
print('Student name:', student['name'])
print('Course:', student.get('course'))

# Modify and add key-value pairs
student['age'] = 22
student['city'] = 'Bengaluru'
print('Updated dictionary:', student)
```

```python
# Iterate through dictionary
print('Dictionary items:')
for key, value in student.items():
    print(f'{key}: {value}')
```

## 4) Choosing the right data structure

- Use a **list** when values may change and order matters.
- Use a **tuple** when values should stay fixed (immutability).
- Use a **dictionary** when data is best represented as named fields (key-value pairs).

```python
summary = {
    'list_use_case': 'Dynamic ordered items',
    'tuple_use_case': 'Fixed record that should not change',
    'dict_use_case': 'Structured entity with named attributes'
}
summary
```
