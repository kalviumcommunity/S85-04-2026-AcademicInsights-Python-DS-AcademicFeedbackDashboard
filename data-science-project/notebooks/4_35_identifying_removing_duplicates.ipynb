# 4.35 — Identifying and Removing Duplicate Records

This notebook demonstrates how to detect, inspect, remove, and verify duplicates in a Pandas DataFrame.

## 1) Load a DataFrame with potential duplicates

```python
import pandas as pd

df = pd.DataFrame([
    {'student_id': 101, 'name': 'Aarav', 'course': 'Python', 'score': 88},
    {'student_id': 102, 'name': 'Diya',  'course': 'Python', 'score': 91},
    {'student_id': 101, 'name': 'Aarav', 'course': 'Python', 'score': 88},  # exact duplicate
    {'student_id': 103, 'name': 'Ishaan','course': 'SQL',    'score': 85},
    {'student_id': 104, 'name': 'Meera', 'course': 'Excel',  'score': 79},
    {'student_id': 102, 'name': 'Diya',  'course': 'Python', 'score': 91},  # exact duplicate
    {'student_id': 105, 'name': 'Riya',  'course': 'Python', 'score': 75},
    {'student_id': 105, 'name': 'Riya',  'course': 'Python', 'score': 83}   # partial duplicate on id+course
])

print('Original shape:', df.shape)
df
```

## 2) Detect duplicate rows

```python
# Boolean indicator for full-row duplicates
full_dup_mask = df.duplicated()

print('Number of full-row duplicates:', full_dup_mask.sum())
print('Duplicate mask:')
print(full_dup_mask.tolist())

print('\nDuplicate rows (all columns):')
df[full_dup_mask]
```

## 3) Detect duplicates using selected columns (partial duplicates)

```python
# Check duplicates by business key: student_id + course
subset_cols = ['student_id', 'course']
partial_dup_mask = df.duplicated(subset=subset_cols, keep=False)

print('Rows involved in partial duplicates (student_id + course):', partial_dup_mask.sum())
df[partial_dup_mask].sort_values(subset_cols)
```

## 4) Remove duplicates intentionally

```python
# A) Remove exact full-row duplicates (keep first occurrence)
df_no_full_dups = df.drop_duplicates()
print('Shape after removing full-row duplicates:', df_no_full_dups.shape)

# B) Remove duplicates by selected columns, keeping the last record
# (useful when latest record should be retained)
df_no_partial_dups = df.drop_duplicates(subset=['student_id', 'course'], keep='last')
print('Shape after removing duplicates by student_id+course (keep last):', df_no_partial_dups.shape)

df_no_partial_dups.sort_values(['student_id', 'course'])
```

## 5) Verify deduplication results

```python
print('Before:', df.shape)
print('After (full-row dedup):', df_no_full_dups.shape)
print('After (subset dedup):', df_no_partial_dups.shape)

remaining_full_dups = df_no_full_dups.duplicated().sum()
remaining_subset_dups = df_no_partial_dups.duplicated(subset=['student_id', 'course']).sum()

print('Remaining full-row duplicates:', remaining_full_dups)
print('Remaining subset duplicates (student_id+course):', remaining_subset_dups)
```

## Conclusion
- Duplicate detection must happen before removal.
- Use full-row dedup for exact repeats.
- Use subset-based dedup when business keys define uniqueness.
- Always verify shape changes and remaining duplicate counts.
